```python
def load_df(fpath, sep='\t', print_log=True, index_col=0, **kwargs):
    if not index_col is None:
        df = pd.read_csv(fpath, sep=sep, index_col=[index_col], **kwargs)
    else:
        df = pd.read_csv(fpath, sep=sep, **kwargs)
    if print_log:
        print(df.shape)
        print(df.columns)
        display(df.head())
    return df
```
